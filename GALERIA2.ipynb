Primer intento para generar una imagen con datos normales multivariados y combinaciones trigonometricas:

```python
import random
from samila import Projection
random.seed(425449)
#%% Funcion trigonometrica 2
import random
import math
def f1(x, y):
    result = random.uniform(-100,100) * x**2 * x1
    return result
def f2(x, y):
    result = random.uniform(-100,100) * (x*x2+y*y2)
    return result
g = GenerativeImage(f1, f2)
g.generate()
g.plot(color="red", bgcolor="black", projection=Projection.POLAR)
#g.seed

plt.show()
```

```python
#%% Funcion trigonometrica 5
import random
import math
random.seed(10)
def f1(x, y):
    result = random.uniform(-1,1) * x*2 - math.sin(y*2) + abs(y-x)
    return result
def f2(x, y):
    result = random.uniform(-1,1) * y*2 - math.cosh(x*2) + 2*math.pi
    return result
g = GenerativeImage(f1, f2)
g.generate(seed=10)
g.plot(projection=Projection.POLAR, color="black", bgcolor="navajowhite")
g.seed
g.save_image(file_adr="test2.png", depth=5)
plt.show()
```
